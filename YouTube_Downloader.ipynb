```python

import os
import subprocess

# Memeriksa apakah paket pytube sudah terpasang
try:
    import pytube
except ImportError:
    print("Paket pytube belum terpasang.")
    print("Melakukan instalasi paket pytube...")
    subprocess.check_call(["pip", "install", "pytube"])
    import pytube

from pytube import YouTube

# Memasukkan URL video YouTube yang ingin diunduh
video_url = "https://youtu.be/05nROLhMkkk" #@param {'type': 'string'}

# Mengunduh video dari YouTube dengan audio dan meningkatkan kecepatan pemrosesan
def download_video_with_audio(video_url, quality='480p'):
    try:
        yt = YouTube(video_url)
        video_title = yt.title
        video_file = f"{video_title} ({quality}).mp4"

        # Menentukan path untuk video
        video_path = f"/content/drive/MyDrive/Videos/{video_file}"

        # Memeriksa apakah video sudah ada
        if os.path.exists(video_path):
            print("Video sudah ada di Google Drive.")
            print("Proses pengunduhan dihentikan.")
            return

        # Mendapatkan stream video dengan audio
        video_stream = yt.streams.filter(res=quality, progressive=True).first()

        if video_stream is None:
            print(f"Video dengan kualitas {quality} tidak tersedia.")
            print("Pilih kualitas yang lain.")
            return

        # Mengunduh video dengan audio dan meningkatkan kecepatan pemrosesan
        video_stream.download(output_path='/content/drive/MyDrive/Videos', filename=video_file)

        print("Disimpan di lokasi Google Drive (Videos)")
        print("Judul video:", video_title)
        print("Kualitas video:", quality)
        print("Video berhasil dikirim")
        return yt  # Mengembalikan objek YouTube

    except Exception as e:
        print("Terjadi kesalahan saat mengunduh video:", str(e))

# Menampilkan opsi dropdown untuk memilih kualitas video
quality_options = ['144p', '240p', '360p', '480p', '720p', '1080p', '1440p', '2160p']
selected_quality = '720p' #@param ['144p', '240p', '360p', '480p', '720p', '1080p', '1440p', '2160p']
print("Kualitas video yang dipilih:", selected_quality)

# Memanggil fungsi untuk mengunduh video dengan audio dan kualitas yang dipilih
yt = download_video_with_audio(video_url, selected_quality)
```
